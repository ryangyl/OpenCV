```python
import cv2
import numpy as np
import requests
from io import BytesIO
from IPython.display import display
# URL of the image
url = 'https://docs.opencv.org/4.x/roi.jpg'

# Fetch the image using requests
response = requests.get(url, verify = False)
image_data = BytesIO(response.content)

# Convert the image data to a numpy array
image_array = np.asarray(bytearray(image_data.read()), dtype=np.uint8)

# Decode the numpy array into an image using OpenCV
img = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
display(img)
```

```python
# Display the image (if using a notebook, consider using matplotlib for display)
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
plt.axis('off')
plt.show()
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

url = 'https://docs.opencv.org/4.x/opencv-logo-small.png'
response = requests.get(url, verify = False)
image_data = BytesIO(response.content)

# Convert the image data to a numpy array
image_array = np.asarray(bytearray(image_data.read()), dtype=np.uint8)

# Decode the numpy array into an image using OpenCV
img1 = cv2.imdecode(image_array, cv2.IMREAD_COLOR)
BLUE = [255,0,0]
 
assert img1 is not None, "file could not be read, check with os.path.exists()"
 
replicate = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_REPLICATE)
reflect = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_WRAP)
constant= cv.copyMakeBorder(img1,10,10,10,10,cv.BORDER_CONSTANT,value=BLUE)
plt.subplot(231),plt.imshow(img1,'gray'),plt.title('ORIGINAL')
plt.axis('off')
plt.subplot(232),plt.imshow(replicate,'gray'),plt.title('REPLICATE')
plt.axis('off')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.axis('off')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.axis('off')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.axis('off')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
plt.axis('off')
plt.show()
```
